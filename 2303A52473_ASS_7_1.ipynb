### Assert Tests for `greet()` function

```python
import io
import sys

def test_greet_function():
    # Capture the standard output
    captured_output = io.StringIO()
    sys.stdout = captured_output

    # Call the function that prints
    greet()

    # Restore standard output
    sys.stdout = sys.__stdout__

    # Get the printed value
    output = captured_output.getvalue()

    # --- Assert Test Cases ---

    # Test Case 1: Check if the exact expected string is printed
    expected_output = "Hello, AI Debugging Lab!\n"
    assert output == expected_output, f"Test Case 1 Failed: Expected '{expected_output}', Got '{output}'"
    print(f"Test Case 1 Passed: Output is exactly '{output.strip()}'")

    # Test Case 2: Check if the output starts with 'Hello'
    assert output.startswith("Hello"), f"Test Case 2 Failed: Output does not start with 'Hello'"
    print("Test Case 2 Passed: Output starts with 'Hello'")

    # Test Case 3: Check if the output contains 'AI Debugging Lab!'
    assert "AI Debugging Lab!" in output, f"Test Case 3 Failed: Output does not contain 'AI Debugging Lab!'"
    print("Test 3 Passed: Output contains 'AI Debugging Lab!'")

    print("\nAll greet() function tests for TASK 1 passed successfully!")

# Run the tests
test_greet_function()
```

```python
def add_five(value):
  return value + 5
print(add_five("10"))
```

The TypeError: can only concatenate str (not 'int') to str means that you are trying to add a string, like '10', to an integer, like 5. Python doesn't know how to do that directly. To fix this, you need to convert the string '10' into an actual number (an integer) before you can add 5 to it.

```python
def add_five(value):
  return int(value) + 5
print(add_five("10"))
```

### Solution 1: Type Casting (Converting string to integer)

```python
def add_five_type_casting(value):
  # Convert the input 'value' to an integer before adding 5
  return int(value) + 5

print("--- Testing Type Casting Solution ---")
# Test cases for type casting solution
assert add_five_type_casting("10") == 15, "Type Casting Test 1 Failed: Expected 15 for '10'"
print("Type Casting Test 1 Passed: add_five_type_casting('10') -> 15")

assert add_five_type_casting("0") == 5, "Type Casting Test 2 Failed: Expected 5 for '0'"
print("Type Casting Test 2 Passed: add_five_type_casting('0') -> 5")

assert add_five_type_casting("-5") == 0, "Type Casting Test 3 Failed: Expected 0 for '-5'"
print("Type Casting Test 3 Passed: add_five_type_casting('-5') -> 0")

print("All Type Casting tests passed successfully!")
```

### Solution 2: String Concatenation (Converting integer to string)

```python
def add_five_str_concat(value):
  # Convert the integer 5 to a string before concatenating with 'value'
  return value + str(5)

print("\n--- Testing String Concatenation Solution ---")
# Test cases for string concatenation solution
assert add_five_str_concat("10") == "105", "String Concatenation Test 1 Failed: Expected '105' for '10'"
print("String Concatenation Test 1 Passed: add_five_str_concat('10') -> '105'")

assert add_five_str_concat("abc") == "abc5", "String Concatenation Test 2 Failed: Expected 'abc5' for 'abc'"
print("String Concatenation Test 2 Passed: add_five_str_concat('abc') -> 'abc5'")

assert add_five_str_concat("hello ") == "hello 5", "String Concatenation Test 3 Failed: Expected 'hello 5' for 'hello '"
print("String Concatenation Test 3 Passed: add_five_str_concat('hello ') -> 'hello 5'")

print("All String Concatenation tests passed successfully!")
```
